To remove NaN values which came from the format of the data being from '.csv' format we'll have to convert the data into list and then remove those values.

```python
df_list = df.values.tolist()
```

Since value of NaN is not equal to Nan in python, since it is null.

```python
new_df = []
li = []
```

```python
for b in range(len(df_list)):
    for a in range(len(df_list[b])):
        if df_list[b][a]==df_list[b][a]:
            li.append(df_list[b][a])
    new_df.append(li)
    li = []
```

```python
len(new_df)
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_arr = te.fit(new_df).transform(new_df)
```

```python
data = pd.DataFrame(te_arr, columns=te.columns_)
data.head()
```

1. True = This item is in the transaction.
2. False = This item is not in the transaction.

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(data, min_support=0.7, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head()
```

```python
lil = []
for a in range(len(new_df)):
    for b in range(len(new_df[a])):
        lil.append(new_df[a][b])
```

```python
len(lil)
```

```python
stripped = [s.strip() for s in lil]
```

```python
output = []
for a in stripped:
    if a not in output:
        output.append(a)
```

```python
len(output)
```

```python
output
```

There are 119 unique items in the whole data.

```python
freq = []
leng = []
```

```python
for a in range(len(output)):
    for c in range(len(new_df)):
        if output[a] in new_df[c]:
            leng.append(a)
    freq.append(len(leng))
    leng = []
```

```python
freq_df = pd.DataFrame(data = {'Items':tuple(output), 'Frequency': tuple(freq)})
freq_df
```

##### Calculating Support, Confidence and Lift

```python
le = []
leng1 = []
leng2 = []
supp = []
supp_cnt = []
conf = []
lif = []
length = []
```

```python
for a in range(len(output)):
    for d in range(len(output)):
        if a!=d:
            for c in range(len(new_df)):       
                if (output[a] in new_df[c]) and (output[d] in new_df[c]):
                    le.append(a)
            leng1.append(output[a])
            leng2.append(output[d])
            length.append(len(le))
            s = len(le)/len(new_df)
            supp.append(s)
            scnt = len(le)
            supp_cnt.append(scnt)
            c = len(le)/freq_df['Frequency'][d]
            conf.append(c)
            l = c/(freq_df['Frequency'][a]/len(new_df))
            lif.append(l)
            le = []
```

```python
print(len(le))
print(len(leng1))
print(len(leng2))
print(len(supp))
print(len(supp_cnt))
print(len(conf))
print(len(lif))
```

```python
rul = pd.DataFrame(data = {'Antecedent': leng1, 'Consequent': leng2, 'Support_count': supp_cnt,\
                           'Support': supp,'Confidence': conf, 'Lift': lif})
rul
```

```python
rul['Support_count'].describe(percentiles=[0.5,0.75,0.77,0.8,0.9,0.95,0.999])
```

```python
new_rul=rul[rul['Support_count']>=10].reset_index(drop=True)
new_rul
```
